# Cleaning Data

```python
import numpy as np
sales_data = {"items":["clothes","footware","jwellery","electronics","cosmetics"],
             "Person": ["Rajesh","Raju", "Pradeep", "Amit", "Rakesh"], 
            "Company" : [ "CompA","CompA", "CompA", "CompB", "CompB"],
             "Sales"  : [   250  ,  250  ,   150  ,   340  ,   np.nan],
             "Age"    : [    23  ,  24   ,   25   ,   250  ,  np.nan] }
sales_df = pd.DataFrame(sales_data)
sales_df
```

```python
sales_df.drop_duplicates(inplace=True)
```

```python
sales_df
```

```python
sales_df["Age"].replace(250.0, 25.0, inplace=True)
sales_df
```

```python
sales_df.dropna()
```

```python
sales_df
```

```python
int(sales_df["Age"].mean())
```

```python
sales_df["Age"].fillna(int(sales_df["Age"].mean()), inplace=True)
```

```python
sales_df
```

```python
sales_df["Sales"].fillna(0, inplace=True)
sales_df
```
